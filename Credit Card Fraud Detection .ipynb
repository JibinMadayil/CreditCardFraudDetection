# Splitting Dataset

```python
X = df.drop(columns=['Class'],axis=1)
y=df['Class']
```

# Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_scaler=sc.fit_transform(X)
```

```python
x_scaler[-1]
```

# Model Training

```python
# train test split
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score
x_train, x_test, y_train, y_test = train_test_split(x_scaler,y,test_size=0.25, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
#model training
log_model.fit(x_train, y_train)
#testing
y_pred=log_model.predict(x_test)
print(classification_report(y_test,y_pred))
print("F1 Score:",f1_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC_model=RandomForestClassifier()
#model training
RFC_model.fit(x_train, y_train)
#testing
y_pred=RFC_model.predict(x_test)
print(classification_report(y_test,y_pred))
print("F1 Score:",f1_score(y_test,y_pred))
```

```python
from xgboost import XGBClassifier
XGB_model=XGBClassifier()
#model training
XGB_model.fit(x_train, y_train)
#testing
y_pred=XGB_model.predict(x_test)
print(classification_report(y_test,y_pred))
print("F1 Score:",f1_score(y_test,y_pred))
```

# Class imbalancement

```python
# balance the class with equal distribution
from imblearn.over_sampling import SMOTE
over_sample=SMOTE()
x_somte,y_smote=over_sample.fit_resample(x_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
#model training
log_model.fit(x_somte,y_smote)
#testing
y_pred=log_model.predict(x_test)
print(classification_report(y_test,y_pred))
print("F1 Score:",f1_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC_model=RandomForestClassifier()
#model training
RFC_model.fit(x_somte,y_smote)
#testing
y_pred=RFC_model.predict(x_test)
print(classification_report(y_test,y_pred))
print("F1 Score:",f1_score(y_test,y_pred))
```

```python
from xgboost import XGBClassifier
XGB_model=XGBClassifier()
#model training
XGB_model.fit(x_somte,y_smote)
#testing
y_pred=XGB_model.predict(x_test)
print(classification_report(y_test,y_pred))
print("F1 Score:",f1_score(y_test,y_pred))
```
